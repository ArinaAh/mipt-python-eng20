# Привет!
Jupyter Notebook – это среда, позволяющая удобно создавать и выполнять (относительно) небольшие программы на Python, а также сохранять и демонстрировать результаты выполнения этих программ. “Единицей хранения” Jupyter Notebook является “блокнот” (собственно, notebook, файл.ipynb). Блокнот состоит из ячеек, они бывают разных типов. Наиболее интересный нам тип ячеек содержит (а) исходный код программы (можно узнать по "In [x]:", где "x" показывает, сколько раз код ячейки был запущен на выполнение) и (б) вывод программы (то, что программа “печатает”, можно узнать по "Out [x]:")

## Рабочая страница блокнота.
На странице блокнота, помимо основного меню со стандартными вкладками “File”, “Edit”, “View” и т.д., есть панель инструментов, которая содержит следующие команды: 
- сохранить (дискета); 
- вставить пустую ячейку (+);
- удалить текущую ячейку (ножницы);
- копировать ячейку в буфер;
- вставить ячейку из буфера; 
- переместить выбранную ячейку (стрелки вверх и вниз); 
- выполнить код в ячейке (кнопка “play” или "Run"); 
- остановить выполнение (кнопка stop);
- перезагрузиться (вам не нужно).

Чтобы запустить код ячейки на выполнение, нужно сделать её активной (кликнуть на её код) и нажать на кнопку “play”. После этого активной станет следующая ячейка!

# 3. Переменные и основные типы данных

Чтобы вы не пугались, давайте введём несколько понятий, которые нам дальше понадобятся:

**Тип** (или **Класс**) - вид данных, с которым мы работаем, например: числа и строки - это типы данных.

**Объект** - экземпляр какого-либо типа (класса), например: число 3 - объект типа "целое число".

**Переменная** - имя, по которому мы можем обращаться к объектам.

**Функция** - процедура, выполняющая какие-либо действия. После имени функции обязательно следуют круглые скобки, внутри которых перечисляются аргументы функции. Например: **ЗамеситьТесто(яйца = 2, мука = 0.5кг, соль = 1ч.л.)**

Стоит выделить два особых вида функций:

1) **Оператор** - это такой элемент языка программирования. Можно сказать, что это некая "встроенная" функция. Операторы с виду на функции не похожи и как правило отделяются от других слов пробелами. Посмотрим на оператор присваивания (привязывания):

````python
x = 3      # Привязываем к переменной x объект типа "целое число" со значением 3 
           #(значок "равно" и есть оператор)

x          # Чтобы jupyter показал нам значение переменной, просто пишем её имя в отдельной строке         
````

2) **Метод** - это функция, созданная специально для какого-либо типа (класса). Вызов метода выглядит так:

````python
result = 'hello'.swapcase() # У строк есть метод "swapcase", который обращает маленькие буквы в большие и наоборот
#Помимо аргументов в круглых скобках, метод принимает ещё один аргумент - значение, "для которого он вызывается" - 
# это объект, который указывается перед вызовом метода и отделяется от него точкой. 
#Т.е. в данном случае метод swapcase принимает объект типа "строка" со значением "hello" и возвращает
#НОВЫЙ объект типа строка со значением "HELLO".
result
````

# 5. Условия и циклы


Результаты сравнений переменных принимают значения типа bool. 

В Python существуют следующие операторы сравнения:
- **==** - проверка равенства
- **\>** - больше
- **<** - меньше
- **\>=** - больше либо равно
- **<=** - меньше либо равно
- **!=** - проверка НЕравенства

Смотрите сами:

````python
5 == 7
8 > 0
-10 <= 9                 
'Курица' == 'Птица'
'делать мне' > 'Нечего'  # Как может работать сравнение строк?
'Делать мне' > 'Нечего'  # Как может работать сравнение строк?
'Лекция' != 'Семинар'
3.0 == 3                 # Python может даже сравнивать разные типы данных!
````

````python
# Булева алгебра

# Есть ещё три оператора:
# not - логическое отрицание (превращает истину в ложь и наоборот)
# and - логическое "И"       (возвращает истину только для двух истин)
# or  - логическое "ИЛИ"     (возвращает ложь только для двух лжей)

# not аналогичен знаку "минус", and - умножению, or - сложению
# Поэтому приоритет операций такой: сначала not, потом and, потом or
# Например: not a and b or c будет выполняться вот так: ((not a) and b) or c
# т.е. сначала "не а", потом "и б", потом "или ц"


istina = True
lozh   = False

res = not istina
res

res = istina and lozh
res

res = istina or lozh
res
````

````python
# Условная конструкция if..elif..else

question = 'Can I drive a car?'

age = int(input("Укажите свой возраст: ")) # Удобно использовать int(input()), если вы запрашиваете число

country = 'Russia'
#country = 'USA'

# Будет выполнен только один из трех блоков кода - тот, для которого будет истинно условие
# Ну или самый последний (иначе), если во всех проверках условия ложны

if age >= 16 and country == 'USA': # Если вам есть 16 И вы живете в США
    
    # блок 1
    answer = 'Yes, you can drive'

elif age >= 18 and country == 'Russia': # Если условие выше неверно, но вам есть 18 И вы живете в России
    
    # блок 2
    answer = 'Yes, you can drive'
    
else: # Иначе
    
    # блок 3
    answer = "No, you can't drive"
    
print(answer)
````

### Замечание 4: 

В Питоне нет операторных скобок, в отличие от таких языков, как С++ или Java. В этих языках набор команд, которые должны быть выполнены при определённом условии, определяется расположением скобок, например что-то вроде:
```
ЕСЛИ условие1, ТО {
команда1
команда2}
ИНАЧЕ {
команда3}
```
Здесь команды 1 и 2 выполняются при условии1, иначе выполняется команда3.

На питоне это выглядит вот так:
```
if условие1:
    команда1
    команда2
else:
    команда3
```

````python
# Есть ещё оператор проверки вхождения in:

if 'I' in 'TEAM':
    pass          # pass - слово-заглушка, т.е. команда "не делать ничего", нужна для пустых блоков
else:
    print('нет буквы я в слове команда')
````

````python
# Цикл for == Цикл ДЛЯ - повторять операции для каждого элемента/индекса

for i in '123':
    print('повтор')
````

````python

slovo = 'dlinnosheee'

# Итерации по элементам строки:

for bukva in slovo:  # Для каждого символа bukva строки из переменной slovo
    print(bukva)
    
    
# Итерации по индексам строки (нумерация с нуля)     индекс == номер элемента в строке
# range(N) - диапазон чисел от 0 до N-1
    
for index in range(len(slovo)): # Для каждого индекса от 0 до длины слова минус 1
    print(index, slovo[index])  # Квадратные скобки позволяют получить элемент с соответствующим индексом
    
# Функция enumerate() позволяет итерироваться по парам индекс-элемент

for index,bukva in enumerate(slovo):
    print(index, bukva)
````

````python
# Слово break - выйти из цикла 

for bukva in 'paparapam':
    
    print(bukva)
    
    if bukva == 'r':    # Если попалась буква r - выходим из цикла сразу
        break

print('вышли из цикла')
````

````python
# Слово continue - перейти к следующей итерации

#Печатаем всё кроме гласных букв

result = ''

for bukva in "дайте мне ещё этих мягких французских булочек":
    
    if bukva in 'уеыаоэяию':  # если буква гласная - прыгаем к следующей
        continue             
        
    result += bukva  # Дописываем букву к результату

print(result)
````

# 6. Списки, множества, словари

````python
# Есть четыре основных типа контейнеров 

list()      # Список. list() - создает пустой список, list('строка') - превращает строку в список
            # Элементами списка могут быть объекты любого типа
            # Список выделяется квадратными скобками
        
[1, True, 3.5, 'privet', [1,2]]  # Даже список может быть элементом списка

tuple()     # Кортеж. Он как список, но неизменяемый (нельзя менять например элементы по индексу)

(1, True, 3.5, 'privet', [1,2])  # Кортеж выделяется круглыми скобками.
 
dict()      # Словарь. Выделяется фигурными скобками. Хранит пары ключ-значение. 
            # Ключ - это как индекс, только ключ может быть объектом (почти) любого типа

d = {'a':1, 'б':2}  # Например строка может быть ключом
d['a']              # Доступ по ключу аналогичен доступу по индексу
d[(1,2)] = 32       # Создаем новую пару ключ-значение - (1,2)-32
d
                   

set()       # Множество - неупорядоченный набор уникальных элементов
            # Как и словарь выделяется фигурными скобками, но хранит только уникальные значения, а не пары

set([1,2,2,3,4,4]) # Делаем множество из списка

# ВАЖНО!
# Кортеж может быть ключом в словаре или элементом множества,
# а список - нет. и сам словарь - нет, и множество - нет
# Общий принцип - изменяемые типы не могут быть ключом, неизменяемые - могут
# Изменяемые - list,dict,set, неизменяемые - все остальные (и строки и числа и пр.)
````

Чтобы получить часть списка нужно использовать конструкцию list_name[a:b], которая выведет элементы списка с индексами от a до b-1. 

Если индекс "a" не указан, то результат будет начинаться с элемента под индексом 0; если не указан индекс "b" – результат будет заканчиваться последним элементом списка.

Также, можно использовать отрицательные индексы. Индекс -1 соответствует последнему элементу списка, -2 – предпоследнему и так далее.

````python
# Срезы
tea_party = ['Rabbit', 'Bear', 'Fox', 'Turtle']

#tea_party[a:b:s]
tea_party[:] # Если нет ни a, ни b - выводится список целиком
tea_party[1:3] # а включительно, b НЕ включительно
tea_party[:2]
tea_party[3:]
tea_party[-1] # Последний элемент списка
tea_party[:-1] # Все элементы списка, кроме последнего
tea_party[0:5:2] # а ещё можно задать шаг прохода s - в данном случае элементы выводятся "через один"
tea_party[::-1] # а если задать шаг равным -1, список можно развернуть!
tea_party[4] # вылезли за пределы
````

````python
# Часто используются списки, элементами которых являются числа, идущие с определенным шагом. 
# Для создания таких списков можно использовать специальную функцию range(start, end, step):
# Результат функции range() не включает значение end!!

# Вызов функции range(N) создает список с элементами от 0 до N-1.
# Если честно - функция range(N) создаёт объект типа "итератор"
# Для наглядности будем преобразовывать итератор в список функцией list()

list(range(10)) # если параметр только один, то это параметр end, а start==0 и step==1
list(range(1,10)) # если параметра два, то это (start, end), а step==1
list(range(0, 20, 3))    # Идем с шагом 3, т.е. берём каждое третье число начиная с 0 и до 20 (не включительно)
list(range(100, 75, -2)) # Шаг может быть и отрицательным
list(range(1,10,0.5)) # А вот нецелым числом шаг быть не может!
````

````python
# Списковые выражения - удобный способ создавать списки (и не только)

[x**2 for x in range(10)]             # Квадраты чисел до 10
[x**2 for x in range(10) if x%2 == 0] # Квадраты чётных чисел до 10

{x: x+1 for x in range(10)} # Словари тоже можно так создавать
{x // 2 for x in range(10)} # И множества 
````

````python
# Минимум, максимум, сумма

min(range(10))
max(range(10))
sum(range(10))
````

````python
# Крайне полезные методы str1.split(str2) и str.join(list):

# str1.split(str2) - разделяет строку str1 на элементы списка, используя str2 как разделитель
# Например с помощью split можно разбить строку-предложение на список слов, используя пробел как разделитель

# str.join(list) - делает обратную операцию, т.е. склеивает элементы списка list в строку, 
#                                            разделяя их значениями str

A = 'Привет, я обычное предложение, во мне есть слова и пробелы'.split(' ')
print(A)

B = ' '.join(A)
print(B)

C = B.split('е')
print(C)

D = 'э'.join(C)
print(D)
````

````python
# Итерации по словарю

dictionary = {'a': 1, 'b': 2, 'c': 3}
   
for k in dictionary.keys():
    print(k)
    
print()
    
for k in dictionary:  # такая же итерация по ключам, но Python Zen говорит нам, что явное лучше, чем неявное
    print(k)          # поэтому лучше явно прописать .keys(), чтобы улучшить читабельность кода
                      # слишком читабельный код еще никогда никому не мешал

print()
    
for v in dictionary.values(): # итерация по значениям
    print(v)
    
print()
    
for pair in dictionary.items(): # итерируемся сразу по парам (ключ, значение)
    print(pair)
    
print()

for key, value in dictionary.items(): # итерируемся сразу по парам (ключ, значение), но с двумя переменными
    print(key, value)
````
